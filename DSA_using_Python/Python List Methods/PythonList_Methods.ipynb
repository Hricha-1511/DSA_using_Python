# 📚 Must-Know Built-in Methods for Python List

Python lists provide several built-in methods for adding, removing, searching, sorting, and manipulating elements efficiently.

---

## 1. `append()`

```python
list.append(value)
```

Adds an item to the **end** of the list.

**Example**

```python
numbers = [10, 20, 30]
numbers.append(40)

print(numbers)
# Output: [10, 20, 30, 40]
```

---

## 2. `extend()`

```python
list.extend(iterable)
```

Adds **all elements** from another iterable (list, tuple, set, etc.) to the end of the list.

**Example**

```python
numbers = [1, 2]
numbers.extend([3, 4, 5])

print(numbers)
# Output: [1, 2, 3, 4, 5]
```

---

## 3. `insert()`

```python
list.insert(index, value)
```

Inserts an element at the specified index.

**Example**

```python
numbers = [10, 20, 30]
numbers.insert(1, 15)

print(numbers)
# Output: [10, 15, 20, 30]
```

---

## 4. `remove()`

```python
list.remove(value)
```

Removes the **first occurrence** of the specified value.

> **Note:** Raises a `ValueError` if the value is not found.

**Example**

```python
numbers = [10, 20, 30, 20]
numbers.remove(20)

print(numbers)
# Output: [10, 30, 20]
```

---

## 5. `pop()`

```python
list.pop(index=-1)
```

Removes and returns the element at the specified index.

If no index is provided, it removes the **last element**.

> **Note:** Raises an `IndexError` if the list is empty or the index is invalid.

**Example**

```python
numbers = [10, 20, 30]

value = numbers.pop()

print(value)
# Output: 30

print(numbers)
# Output: [10, 20]
```

---

## 6. `clear()`

```python
list.clear()
```

Removes **all elements** from the list.

**Example**

```python
numbers = [1, 2, 3]
numbers.clear()

print(numbers)
# Output: []
```

---

## 7. `index()`

```python
list.index(value, start=0, stop=len(list))
```

Returns the index of the **first occurrence** of the specified value.

> **Note:** Raises a `ValueError` if the value is not found.

**Example**

```python
numbers = [10, 20, 30, 20]

print(numbers.index(20))
# Output: 1
```

---

## 8. `count()`

```python
list.count(value)
```

Returns the number of times a value appears in the list.

**Example**

```python
numbers = [10, 20, 20, 30, 20]

print(numbers.count(20))
# Output: 3
```

---

## 9. `sort()`

```python
list.sort(key=None, reverse=False)
```

Sorts the list **in ascending order** by default.

Use `reverse=True` for descending order.

**Example**

```python
numbers = [50, 10, 30, 20]
numbers.sort()

print(numbers)
# Output: [10, 20, 30, 50]
```

Descending order:

```python
numbers.sort(reverse=True)

print(numbers)
# Output: [50, 30, 20, 10]
```

---

## 10. `reverse()`

```python
list.reverse()
```

Reverses the order of elements **in place**.

**Example**

```python
numbers = [10, 20, 30]
numbers.reverse()

print(numbers)
# Output: [30, 20, 10]
```

---

## 11. `copy()`

```python
list.copy()
```

Creates and returns a **shallow copy** of the list.

**Example**

```python
numbers = [10, 20, 30]

copy_numbers = numbers.copy()

print(copy_numbers)
# Output: [10, 20, 30]
```
