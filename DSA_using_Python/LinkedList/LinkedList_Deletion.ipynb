# Singly Linked List (SLL) Deletion Algorithms

---

## 1. Delete from Beginning

```text
Algorithm: DELETE_FROM_BEGINNING()

IF head == NULL
    PRINT "Linked List is Empty"
    RETURN

head = head.next
```

**Time Complexity:** O(1)  
**Space Complexity:** O(1)

---

## 2. Delete from End

```text
Algorithm: DELETE_FROM_END()

IF head == NULL
    PRINT "Linked List is Empty"
    RETURN

IF head.next == NULL
    head = NULL
    RETURN

prev = NULL
curr = head

WHILE curr.next != NULL
    prev = curr
    curr = curr.next

prev.next = NULL
```

**Time Complexity:** O(n)  
**Space Complexity:** O(1)

---

## 3. Delete at a Given Position

```text
Algorithm: DELETE_AT_POSITION(pos)

IF head == NULL
    PRINT "Linked List is Empty"
    RETURN

IF pos < 0 OR pos >= length
    PRINT "Invalid Position"
    RETURN

IF pos == 0
    head = head.next
    RETURN

prev = NULL
curr = head
count = 0

WHILE count < pos
    prev = curr
    curr = curr.next
    count = count + 1

prev.next = curr.next
```

**Time Complexity:** O(n)  
**Space Complexity:** O(1)

---

## 4. Delete by Value

```text
Algorithm: DELETE_BY_VALUE(value)

IF head == NULL
    PRINT "Linked List is Empty"
    RETURN

IF head.data == value
    head = head.next
    RETURN

prev = head
curr = head.next

WHILE curr != NULL

    IF curr.data == value
        prev.next = curr.next
        RETURN

    prev = curr
    curr = curr.next

PRINT "Value Not Found"
```

**Time Complexity:** O(n)  
**Space Complexity:** O(1)

````python
# Create a Node class
class Node:
  def __init__(self,data):
    self.data = data
    self.next = None

# Create a Linked List class
class LinkedList:
  def __init__(self):
    self.head = None
   # function to return the length of the linked list
  def length(self):
    count = 0
    temp = self.head
    while temp:
      count +=1
      temp = temp.next
    return count
  # Function to Display the Linked List
  def display(self):
    if self.head is None:
      print("The Linked List is Empty")
    else:
      temp = self.head
      while temp:
        print(temp.data, end = " -> ")
        temp = temp.next
      print("None")
  # function to delete the starting node
  def delete_from_beginning(self):
    if self.head is None:
      print("The Linked List is empty")
    else:
      self.head = self.head.next
  # Function to delete the last node of the LL
  def delete_from_end(self):
    if self.head is None:
      print("The Linked List is empty")
      return

      #only one node
    if self.head.next is None:
      self.head = None
      return
    curr = self.head
    prev = self.head
    while curr.next:
      prev = curr
      curr = curr.next
    prev.next = None

  #Function to delete the middle node from the Linked List
  def delete_from_middle(self, pos):
    if self.head is None:
      print("The Linked List is empty")
      return
    count = 0
    curr = self.head
    prev = self.head
    if pos == 0:
      self.head = self.head.next
      return
    if pos >= self.length() or pos < 0:
      print("Invalid Position")
    else:
      while count < pos:
        prev = curr
        curr = curr.next
        count += 1
      prev.next = curr.next

  def delete_by_value(self, value):
      if self.head is None:
        print("The Linked List is empty")
        return
      if self.head.data == value:
        self.head = self.head.next
        return
      curr = self.head
      prev = self.head
      while curr:
        if curr.data == value:
          prev.next = curr.next
          return
        prev = curr
        curr = curr.next
      print("Value Not Found")




````

````python
l1 = LinkedList()

l1.head = Node(1)
l1.head.next = Node(2)
l1.head.next.next = Node(3)
l1.head.next.next.next = Node(4)

l1.display()

l1.delete_from_beginning()
l1.display()

l1.delete_from_end()
l1.display()

l1.delete_from_middle(1)
l1.display()
````
