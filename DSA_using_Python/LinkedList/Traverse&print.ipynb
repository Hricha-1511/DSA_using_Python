# **3. Traverse and Print the LL**

```python

# Create a Node Class
class Node:
  def __init__(self,data):
    self.data = data
    self.next = None
#Create a Linked List class
class LinkedList():
  def __init__(self):
    self.head = None
  def print_ll(self):
    if self.head is None:
      print("Linked List is Empty")
    else:
      temp = self.head
      while temp.next is not None:
        print(temp.data,end="->")
        temp = temp.next
      print("None")
```
