# What is a Linked List


A linked list is a type of linear data structure individual items are not necessarily at contiguous locations. The individual items are called nodes and connected with each other using links.

A node contains two things first is data and second is a link that connects it with another node.

The first node is called the head node and we can traverse the whole list using this head and next links

# Operations on the Linked List
1. Traverse and print
2. Length of the Linked List
3. Insertion
4. Deletion


```python

# insertion at the Beginning  of the linked list

#  We have a LL 4->5->6->7->8->9 and want to insert 3 at the beginning of the LinkedList

# Node class
class Node:
  def __init__(self,data):
    self.data = data
    self.next = None

# LinkedList class
class LinkedList:
  def __init__(self):
    self.head = None

  #Function to print the length of the linked list
  def length(self):
    count = 0
    temp = self.head

    while temp is not None:
        count+=1
        temp = temp.next
    return count

  #Function to print the Linked List
  def display(self):
    temp = self.head
    while temp is not None:
      print(temp.data,end="->")
      temp = temp.next
    print('None')


  #Function to insert at the beginning of the Linked List
  def insert_at_Beginning(self, data):
    newNode = Node(data)
    newNode.next = self.head
    self.head = newNode

  #Function to insert node at the  end of the Linked List
  def insert_at_end(self,data):
    newNode = Node(data)
    temp = self.head
    while temp.next is not None:
      temp = temp.next
    temp.next = newNode

  #Function to insert node at the middle position given pos is k
  def insert_at_middle(self,data,pos):
    newNode = Node(data)
    temp = self.head
    for i in range(pos-1):
      temp = temp.next
    newNode.next = temp.next
    temp.next = newNode






```

Now let's insert 25 at position = 2.

Here, position means after index 1 (or before index 2).

Current indexes:

Index: 0    1    2    3    4
       ↓    ↓    ↓    ↓    ↓
      10 → 20 → 30 → 40 → 50

```python
# Lets create a LinkedList

l1 = LinkedList()
l1.head = Node(4)
l1.head.next = Node(5)
l1.head.next.next = Node(6)
l1.head.next.next.next = Node(7)
l1.head.next.next.next.next = Node(8)
l1.head.next.next.next.next.next = Node(9)


print("Length:", l1.length())   # 6
l1.display()                    # 4->5->6->7->8->9->None

l1.insert_at_Beginning(3)
l1.display()

print("Length:", l1.length())   # 7
l1.display()

l1.insert_at_end(10)
l1.display()

l1.insert_at_middle(0,3)
l1.display()
```
