# In Python, a set is a built-in data structure used to store an unordered collection of unique items.

####  Because sets utilize an underlying hash table implementation, they provide highly efficient O(1) time complexity for lookups, insertions, and deletions.

# Core Characteristics



**Unique Elements**: Duplicate values are automatically discarded.

**Unordered**: Items do not have a fixed index or position.

**Unindexed**: You cannot access items using square brackets like my_set[0].

**Mutable Structure**: You can add or remove items after creation.

Hashable Items: The individual elements inside the set must be immutable (e.g., strings, numbers, or tuples).

Mutable structures like lists or dictionaries cannot be set elements.

# Methods in python

1.add()

2.union()

3.intersection()

4.difference()

5.clear()

```python
#  1.   Add()

s = {1,2,3}
s.add(4)
print(s)
```

```python
# 2. clear()
s = {1, 2, 3}
s.clear()
print(s)
```

```python
# 3. copy()  :Returns a shallow copy of the set.
s = {1, 2, 3}
c = s.copy()
print(c)
```

```python
# 4.  difference(): Returns a set containing elements present in the first set but not in the second set.

a = {1,2,3,4}
b = {3,4,5}

print(a.difference(b))
```

```python
# 5. difference_update() : Removes common elements from the original set.

a = {1, 2, 3, 4}
b = {3, 4, 5}
a.difference_update(b)
print(a)
```

```python
# 6. discard()  : Removes an element from the set if it exists.
s = {1, 2, 3}
s.discard(2)
print(s)
```

```python
# 7. frozenset() : Creates an immutable set.
s = frozenset([1, 2, 3])
print(s)
```

```python
# 8. intersection() :Returns common elements from two or more sets.
```

```python
# 9. intersection_update() : Updates the set with only the elements that are common to both sets.
a = {1, 2, 3}
b = {2, 3, 4}
a.intersection_update(b)
print(a)
```

```python
# 10. isdisjoint() :Returns True if two sets have no common elements.

a = {1, 2}
b = {3, 4}
print(a.isdisjoint(b))
```

```python
#11. issubset(): Returns True if all elements of one set are present in another set.
a = {1, 2}
b = {1, 2, 3, 4}
print(a.issubset(b))
```

```python
#12. issuperset() :  Returns True if a set contains all elements of another set.

a = {1, 2, 3, 4}
b = {1, 2}
print(a.issuperset(b))
```

```python
#13. pop() : Removes and returns a random element from the set.
s = {1, 2, 3}
print(s.pop())
print(s)
```

```python
#14.  remove()  :Removes the specified element from the set.
s = {1, 2, 3}
s.remove(2)
print(s)
```

```python
#15. symmetric_difference()   :Returns elements that are present in either set but not in both.
a = {1, 2, 3}
b = {3, 4, 5}
print(a.symmetric_difference(b))
```

```python
#16. symmetric_difference_update()   :Updates the set with the symmetric difference of two sets.
a = {1, 2, 3}
b = {3, 4, 5}
a.symmetric_difference_update(b)
print(a)
```

```python
#17. union() :Returns a new set containing all unique elements from the sets.
a = {1, 2, 3}
b = {3, 4, 5}
print(a.union(b))
```

```python
# 18. update()  :Adds all elements from another iterable to the set.
a = {1, 2}
b = {3, 4}
a.update(b)
print(a)
```
