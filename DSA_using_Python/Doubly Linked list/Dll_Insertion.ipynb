# Doubly Linked List Insertion Operation:

1. Insertion at the beginning
2. Insertion at the End
3. Insertion at the given pos of the Doubly Linked List

 ***A doubly linked list is a more complex data structure than a singly linked list, but it offers several advantages. The main advantage of a doubly linked list is that it allows for efficient traversal of the list in both directions. This is because each node in the list contains a pointer to the previous node and a pointer to the next node. This allows for quick and easy insertion and deletion of nodes from the list, as well as efficient traversal of the list in both directions.***

If you want 0-based indexing, then:

pos = 0 → Insert at the beginning.


pos = 1 → Insert after the first node.


pos = 2 → Insert after the second node.
and so on.

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None


class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None
    #function to display the Doubly Linked List
    def display(self):
        if self.head is None:
          print("Doubly Linked list is empty")
        else:
          temp = self.head
          while temp is not None:
            print(temp.data, end ="<->")
            temp = temp.next
          print("None")

    # Insertion at the beginning of the Doubly Linked List
    def insert_at_beginning(self,data):
      newnode = Node(data)
      if self.head is None:
          self.head = self.tail = newnode
      else:
          newnode.prev = None
          newnode.next = self.head
          self.head.prev = newnode
          self.head = newnode

    #Function to insert at the end of the Doubly Linked list
    def insert_at_end(self,data):
       newnode = Node(data)
       if self.head is None:
         self.head = self.tail = newnode
       else:
        self.tail.next = newnode
        newnode.prev = self.tail
        self.tail = newnode


    def insert_at_givenpos(self,pos,data):
      newnode = Node(data)
      # Invalid position
      if pos < 0:
        print("Invalid Position")
        return
      elif pos == 0:
        self.insert_at_beginning(data)
      # Traverse to (pos-1)th node
      else:
        temp = self.head
        for i in range(1,pos):
          if temp is None:
            print("Position out of range")
            return
          temp = temp.next
        if temp is None:
          print("Position out of range")
          return
        if temp.next is None:
          self.insert_at_end(data)
        else:
          newnode.next = temp.next
          newnode.prev = temp
          temp.next.prev = newnode
          temp.next = newnode

```

```python
dll = DoublyLinkedList()

first = Node(10)
second = Node(20)

dll.head = first
dll.tail = second

first.next = second
second.prev = first

dll.display()

dll.insert_at_beginning(5)
dll.display()

dll.insert_at_end(25)
dll.display()
# pos 2 means insertion will take place after 2nd node
dll.insert_at_givenpos(2,15)
dll.display()

dll.insert_at_givenpos(0,1)
dll.display()

dll.insert_at_givenpos(6,30)   #will insert after 6th node
dll.display()
```
