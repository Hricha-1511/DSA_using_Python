```python
# Node class
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

# Linked List class
class LinkedList:
    #constructor
    def __init__(self):
        self.head = None
    # Return the sum of all the nodes of the linked list
    def sum_nodes(self,node):
      if node is None:
         return 0
      return node.data + self.sum_nodes(node.next)
    # Print the linked list using recursion
    def print_LL(self, node):
        if node is None:
          return
        print(node.data,end="->")
        self.print_LL(node.next)
    #return the length of the linked list using recursion
    def length(self, node):
      if node is None:
        return 0
      return 1 + self.length(node.next)
    #return the reverse of the linked list
    def reverse(self,node):
      if node is None:
        return

      self.reverse(node.next)
      print(node.data,end="->" )





```

```python
l = [1,2,3,4,5,6,7,8,9]
l = LinkedList()
l.head = Node(1)
l.head.next = Node(2)
l.head.next.next = Node(3)
l.head.next.next.next = Node(4)


print(l.sum_nodes(l.head))

print(l.print_LL(l.head))

print(l.length(l.head))

print(l.reverse(l.head))
```
