# Data Structure in Python :  TUPLE

#  Python tuples are immutable, meaning they cannot be modified after creation.

Because of this immutability, Python tuples have only two built-in methods:

count()  

index()

image.png

# 1. The count() Method

This method searches the tuple and counts every time the specified element appears. It returns 0 if the element is missing.

```python
fruits = ('apple', 'banana', 'apple', 'cherry', 'apple')

# Count occurrences of 'apple'
print(fruits.count('apple'))  # Output: 3
print(fruits.count('orange')) # Output: 0
```

# 2. The index() Method


This method scans the tuple from left to right and returns the first index position where the specified value is found.

```python
numbers = (10, 20, 30, 20, 40, 50)

# Find first occurrence of 20
print(numbers.index(20))  # Output: 1

# Find 20 starting the search from index 2
print(numbers.index(20, 2))  # Output: 3

# Throws ValueError because 100 is absent
# print(numbers.index(100))
```

While tuples only have two methods called via dot notation, you can pass tuples into several global Python built-in functions to process:image.png

```python
data = (5, 1, 8, 3)

print(len(data))     # Output: 4
print(max(data))     # Output: 8
print(sum(data))     # Output: 17
print(sorted(data))  # Output: [1, 3, 5, 8]
```

### Will make a tuple class to understand the methods

```python
class Tuple:
  def __init__(self, *values):
    self.items = tuple(values)
  #count
  def count(self,value):
    c = 0
    for item in self.items:
      if item == value:
        c=c+1
    return c
  #index
  def index(self,value):
    for i in range(len(self.items)):
      if self.items[i]==value:
        return i
    return -1
  #length()
  def length(self):
    return len(self.items)
  #display()
  def display(self):
    print(self.items)
```

```python
tup = Tuple(10,20,30,20,50)
tup.display()

print(tup.count(20))

print(tup.index(30))

print(tup.length())


print(sum(tup.items))

print(max(tup.items))

print(min(tup.items))

print(sorted(tup.items))
```
